```python
import ast

def convert3(obj):
    L = []  # Initialize an empty list
    counter = 0  # Initialize counter

    for i in ast.literal_eval(obj):  # Safely evaluate the string as a Python object
        if counter != 3:
            L.append(i['name'])  # Append the 'name' value to the list
            counter += 1  # Increment counter
        else:
            break  # Exit loop after 3 elements
    
    return L  # Return the list
```

```python
df['cast']=df['cast'].apply(convert3)
```

```python
import ast

def fetch_director(obj):
    L = []  # Initialize an empty list
    
    for i in ast.literal_eval(obj):  # Convert string representation to a list of dictionaries
        if i['job'] == 'Director':  # Check if the job is 'Director'
            L.append(i['name'])  # Append director's name to the list
            break  # Stop after finding the first director
    
    return L  # Return the list

```

```python
df['crew']=df['crew'].apply(fetch_director)
```

```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Step 1: Data Preprocessing
def convert(obj):
    try:
        obj = ast.literal_eval(obj)  # Convert stringified lists into actual lists
        return [i['name'] for i in obj]
    except:
        return []

df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)

# Extract first 3 actors from cast
def extract_cast(obj):
    try:
        obj = ast.literal_eval(obj)
        return [i['name'] for i in obj[:3]]
    except:
        return []

df['cast'] = df['cast'].apply(extract_cast)

# Extract director name from crew
def extract_director(obj):
    try:
        obj = ast.literal_eval(obj)
        for i in obj:
            if i['job'] == 'Director':
                return [i['name']]
        return []
    except:
        return []

df['crew'] = df['crew'].apply(extract_director)

# Convert overview into a list of words
#df['overview'] = df['overview'].fillna("").apply(lambda x: x.split())

df['overview'] = df['overview'].apply(lambda x: " ".join(x) if isinstance(x, list) else x)
df['overview'] = df['overview'].fillna("").apply(lambda x: x.split())  # Now apply split


# Step 2: Feature Engineering
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
df['tags'] = df['tags'].apply(lambda x: " ".join(x))

# Step 3: Convert Tags into Vectors
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()

# Step 4: Compute Cosine Similarity
similarity = cosine_similarity(vectors)

# Step 5: Movie Recommendation Function
def recommend(movie):
    if movie not in df['title'].values:
        print("Movie not found in database!")
        return
    
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    
    # Get top 5 most similar movies (excluding itself)
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"Movies similar to '{movie}':")
    for i in movies_list:
        print(df.iloc[i[0]]['title'])


```

```python
# Example Usage:
recommend("10 Cloverfield Lane")
```
